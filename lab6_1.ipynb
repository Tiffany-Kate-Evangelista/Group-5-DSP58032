```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = np.arange(100) 

# Input signal x[n] = sin(0.2 * pi * n)
x_n = np.sin(0.2 * np.pi * n)

# Initialize output signal y[n], same length as x[n]
y_n = np.zeros(100)

# Apply the difference equation: y[n] = 0.7 * y[n-1] + x[n]
for i in range(0, 100):
    y_n[i] = 0.7 * y_n[i - 1] + x_n[i]

# Plot input signal x[n] and output signal y[n]
plt.figure(figsize=(10, 6))

# Plot x[n]
plt.subplot(3, 1, 1)
plt.plot(n, x_n, label='Input signal x[n]', color='blue')
plt.title('Input Signal x[n] = sin(0.2πn)')
plt.xlabel('Sample index n')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()

# Plot y[n]
plt.subplot(3, 1, 2)
plt.plot(n, y_n, label='Output signal y[n]', color='red')
plt.title('Output Signal y[n] (System Response)')
plt.xlabel('Sample index n')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()

# Plot input and output signals together
plt.subplot(3, 1, 3)
plt.plot(n, x_n, label='Input signal x[n]', color='blue')
plt.plot(n, y_n, label='Output signal y[n]', color='red')
plt.title('Input and Output Signals')
plt.xlabel('Sample index n')
plt.ylabel('Amplitude')
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```
