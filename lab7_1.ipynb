```python
import numpy as np
import matplotlib.pyplot  as plt
```

```python
# Step 1: Generate the noisy input signal
n = np.arange(100)  # n ranges from 0 to 99
f = 0.1  # Frequency of the sinusoidal signal
clean_signal = np.sin(2 * np.pi * f * n)  # Clean sinusoidal signal
noise = np.random.normal(0, 0.5, len(n))  # Random noise (mean 0, std dev 0.5)
noisy_signal = clean_signal + noise  # Noisy signal

# Step 2: Apply a causal moving average filter
filtered_signal = np.zeros(len(n))  # Initialize filtered signal
for i in range(2, len(n)):
    filtered_signal[i] = (noisy_signal[i] + noisy_signal[i-1] + noisy_signal[i-2]) / 3

# Step 3: Plot the results
plt.figure(figsize=(10, 6))
plt.plot(n, noisy_signal, label='Noisy Signal', linestyle='--', color='r')
plt.plot(n, filtered_signal, label='Filtered Signal', linestyle='-', color='b')
plt.title('Noisy Signal and Filtered Signal using Causal Moving Average Filter')
plt.xlabel('Sample Index')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()
plt.show()
```
