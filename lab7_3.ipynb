```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = np.arange(100)               
x = np.sin(0.1 * np.pi * n)      
d = 5                           
alpha = 0.6                     

x_echo = np.zeros_like(x)
x_echo[d:] = alpha * x[:-d]       
x_with_echo = x + x_echo         
y = x_with_echo - alpha * np.pad(x[:-d], (d, 0), 'constant')

plt.figure(figsize=(10, 6))
plt.plot(n, x_with_echo, label='Signal with Echo', color='blue')
plt.plot(n, y, label='Echo-Canceled Signal', color='red')
plt.title('Echo Cancellation in Telecommunications')
plt.xlabel('Sample number (n)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid()
plt.show()
```
