```python
import numpy as np
import matplotlib.pyplot  as plt
```

```python
t = np.linspace(0, 10, 1000)
x_t = np.sin(t)
y_t = np.gradient(x_t, t)
tau = 2

t_shifted = t - tau
x_t_shifted = np.sin(t_shifted)
y_t_shifted = np.gradient(x_t_shifted, t)
plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(t, y_t, label="y(t) = d/dt*x(t)" , color='green')
plt.xlabel("X")

plt.plot(t, y_t_shifted, label="y(t - tau) = d/dt*x(t - tau)", color='red')
plt.title("Original and Time-Shifted Signal (Differentiation)")
plt.ylabel("Y")
plt.legend()
plt.grid()
plt.show()
```
