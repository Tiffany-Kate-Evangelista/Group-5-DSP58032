```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_samples = 100
base_price = 100
noise = np.random.normal(0, 50, n_samples)  # Random noise
stock_prices = base_price + noise  # Stock prices fluctuating around 100

smoothed_prices = np.zeros(n_samples)

for n in range(1, n_samples - 1):
    smoothed_prices[n] = (stock_prices[n - 1] + stock_prices[n] + stock_prices[n + 1]) / 3

plt.figure(figsize=(10, 6))
plt.plot(stock_prices, label='Original Stock Prices', color='blue', alpha=0.7)
plt.plot(smoothed_prices, label='Smoothed Stock Prices (Non-Causal Filter)', color='red')
plt.title('Stock Price Prediction (Non-Causal Moving Average Filter)')
plt.xlabel('Time')
plt.ylabel('Price')
plt.legend()
plt.grid()
plt.show()
```
