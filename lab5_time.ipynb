```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 10, 1000)  # Time vector from 0 to 10 with 1000 samples
tau = 2                      

x_t = np.sin(t)                
y_t = t * x_t                  

t_shifted = t - tau            
x_t_shifted = np.sin(t_shifted)  
y_t_shifted = (t_shifted) * x_t_shifted  


plt.figure(figsize=(10,6))


plt.subplot(2, 1, 1)
plt.plot(t, y_t, label='y(t) = t * x(t)', color = 'blue')
plt.plot(t, y_t_shifted, label='y(t-τ) = (t-τ) * x(t-τ)', color='red')
plt.title('y(t)')
plt.xlabel('Time (t)')
plt.ylabel('y(t)')
plt.grid()
plt.legend()

# Plot shifted y(t-τ)
#plt.subplot(2, 1, 2)

#plt.title('Shifted System Output y(t-τ)')
#plt.xlabel('Time (t)')
plt.ylabel('y(t-τ)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 1. Time Vector Definition
t = np.linspace(0, 10, 1000)  # Time vector from 0 to 10 with 1000 samples
dt = t[1] - t[0]              # Time step (difference between consecutive time points)
tau = 2                        # Time shift τ

# 2. Position x(t) = sin(t)
x_t = np.sin(t)                # Car's position as a function of time (sine wave)

# 2. Velocity: v(t) = dx/dt
v_t = np.gradient(x_t, dt)     # Velocity as the derivative of x(t)

# 3. Shifting the Input: x(t-τ) and v(t-τ) = d/dt * x(t-τ)
t_shifted = t - tau            # Shifted time vector
x_t_shifted = np.sin(t_shifted)  # Shifted position x(t-τ)
v_t_shifted = np.gradient(x_t_shifted, dt)  # Shifted velocity v(t-τ) = d/dt * x(t-τ)

# Plotting the results
plt.figure(figsize=(10,6))

# Plot original velocity v(t)
plt.subplot(2, 1, 1)
plt.plot(t, v_t, label="v(t) = d/dt * sin(t)")
plt.title('Original System Output v(t)')
plt.xlabel('Time (t)')
plt.ylabel('v(t)')
plt.grid(True)
plt.legend()

# Plot shifted velocity v(t-τ)
plt.subplot(2, 1, 2)
plt.plot(t, v_t_shifted, label="v(t-τ) = d/dt * sin(t-τ)", color='orange')
plt.title('Shifted System Output v(t-τ)')
plt.xlabel('Time (t)')
plt.ylabel('v(t-τ)')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
