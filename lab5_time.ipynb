```python
import numpy as np
import matplotlib.pyplot  as plt
```

```python
t = np.linspace(0, 10, 1000)  # Time vector from 0 to 10 with 1000 samples
tau = 2                      

x_t = np.sin(t)                
y_t = t * x_t                  

t_shifted = t - tau            
x_t_shifted = np.sin(t_shifted)  
y_t_shifted = (t_shifted) * x_t_shifted  


plt.figure(figsize=(10,6))


plt.subplot(2, 1, 1)
plt.plot(t, y_t, label='y(t) = t * x(t)', color = 'blue')
plt.plot(t, y_t_shifted, label='y(t-τ) = (t-τ) * x(t-τ)', color='red')
plt.title('y(t)')
plt.xlabel('Time (t)')
plt.ylabel('y(t)')
plt.grid()
plt.legend()

# Plot shifted y(t-τ)
#plt.subplot(2, 1, 2)

#plt.title('Shifted System Output y(t-τ)')
#plt.xlabel('Time (t)')
plt.ylabel('y(t-τ)')
plt.legend()
plt.tight_layout()
plt.show()
```
