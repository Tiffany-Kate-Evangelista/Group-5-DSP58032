```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Belarmino
# Time vector: 1000 samples from 0 to 10 seconds
t = np.linspace(0, 10, 1000)

# Define the input signal as a sine wave
x_t = np.sin(45 * t) 

# Amplified output: y(t) = 2 * x(t)
y_t = 2 * x_t

# Shifted input: x(t - 2)
tau = 2
x_shifted = np.sin(t - tau)

# Output for shifted input: y(t - 2) = 2 * x(t - 2)
y_shifted = 2 * x_shifted

# Plotting the signals
plt.figure(figsize=(10, 6))

# Plot original input signal x(t)
plt.plot(t, x_t, label='x(t): Original Input', color='b')

# Plot amplified output y(t)
plt.plot(t, y_t, label='y(t) = 2 * x(t): Amplified Output', color='r')

# Plot shifted and amplified signal y(t - 2)
plt.plot(t, y_shifted, label='y(t - tau) = 2 * x(t - tau): Amplified and Shifted Output', color='g')

plt.title('Time-Invariant System (Multiplication by a Constant)')
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()
```
