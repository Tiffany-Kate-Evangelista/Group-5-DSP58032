```python
#define emojis
happy, laughing, sad = "\U0001F600", "\U0001F606", "\U0001F62D"

#calculate semestral grade
def calculate_semestral_grade(prelim, midterm, final):
    return (prelim + midterm + final) / 3

#calculate grade point
def calculate_grade_point(grade):
    if grade >= 97:
        return 1.00
    elif grade >= 93:
        return 1.25
    elif grade >= 89:
        return 1.50
    elif grade >= 85:
        return 1.75
    elif grade >= 82:
        return 2.00
    elif grade >= 79:
        return 2.25
    elif grade >= 76:
        return 2.50
    elif grade >= 73:
        return 2.75
    elif grade >= 70:
        return 3.00
    else:
        return 5.00

#calculate letter grade
def calculate_letter_grade(grade):
    if grade >= 97:
        return "A"
    elif grade >= 93:
        return "A-"
    elif grade >= 89:
        return "B+"
    elif grade >= 85:
        return "B"
    elif grade >= 82:
        return "B-"
    elif grade >= 79:
        return "C+"
    elif grade >= 76:
        return "C"
    elif grade >= 73:
        return "C-"
    elif grade >= 70:
        return "D"
    else:
        return "F"

#main
name = input("Enter your name: ")
course = input("Enter the course name: ")
prelim = float(input("Enter your prelim grade: "))
midterm = float(input("Enter your midterm grade: "))
final = float(input("Enter your final grade: "))

semestral_grade = calculate_semestral_grade(prelim, midterm, final)
grade_point = calculate_grade_point(semestral_grade)
letter_grade = calculate_letter_grade(semestral_grade)

#put emoji
if semestral_grade > 70.00:
    emoji = happy
elif semestral_grade == 70.00:
    emoji = laughing
else:
    emoji = sad

#result
print(f"Name: {name}")
print(f"Course: {course}")
print(f"Semestral Grade: {semestral_grade:.2f} {emoji}")
print(f"Grade Point: {grade_point:.2f}")
print(f"Letter Grade: {letter_grade}")
```
