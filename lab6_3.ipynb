```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
n_samples = 100  # Number of samples
n = np.arange(n_samples)  # Sample indices

# Clean input signal: x_clean[n] = sin(0.3 * pi * n)
x_clean = np.sin(0.3 * np.pi * n)

# Additive white Gaussian noise
noise = np.random.normal(0, 0.5, n_samples)

# Noisy input signal: x_noisy[n] = x_clean[n] + noise
x_noisy = x_clean + noise

# Initialize the output signal y[n]
y = np.zeros(n_samples)

# Difference equation: y[n] = 0.5 * y[n-1] + 0.5 * x_noisy[n]
for i in range(1, n_samples):
    y[i] = 0.5 * y[i - 1] + 0.5 * x_noisy[i]

# Plotting the results
plt.figure(figsize=(10, 6))

# Plot clean input signal
plt.subplot(3, 1, 1)
plt.plot(n, x_clean, label='Clean Input Signal', color='red')
plt.title('Clean Input Signal')
plt.xlabel('n (samples)')
plt.ylabel('Amplitude')
plt.grid(True)

# Plot noisy input signal
plt.subplot(3, 1, 2)
plt.plot(n, x_noisy, label='Noisy Input Signal', color='orange')
plt.title('Noisy Input Signal')
plt.xlabel('n (samples)')
plt.ylabel('Amplitude')
plt.grid(True)

# Plot system output
plt.subplot(3, 1, 3)
plt.plot(n, y, label='System Output', color='green')
plt.title('System Output Signal')
plt.xlabel('n (samples)')
plt.ylabel('Amplitude')
plt.grid(True)

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```

```python
# Parameters
n_samples = 100  # Number of samples
n = np.arange(n_samples)  # Sample indices

# Clean input signal: x_clean[n] = sin(0.3 * pi * n)
x_clean = np.sin(0.3 * np.pi * n)

# Additive white Gaussian noise
noise = np.random.normal(0, 0.5, n_samples)

# Noisy input signal: x_noisy[n] = x_clean[n] + noise
x_noisy = x_clean + noise

# Initialize the output signal y[n]
y = np.zeros(n_samples)

# Difference equation: y[n] = 0.5 * y[n-1] + 0.5 * x_noisy[n]
for i in range(1, n_samples):
    y[i] = 0.5 * y[i - 1] + 0.5 * x_noisy[i]

# Plotting the results
plt.figure(figsize=(10, 8))

# Plot clean input signal
plt.subplot(4, 1, 1)
plt.plot(n, x_clean, label='Clean Input Signal', color='red')
plt.title('Clean Input Signal')
plt.xlabel('n (samples)')
plt.ylabel('Amplitude')
plt.grid(True)

# Plot noisy input signal
plt.subplot(4, 1, 2)
plt.plot(n, x_noisy, label='Noisy Input Signal', color='green')
plt.title('Noisy Input Signal')
plt.xlabel('n (samples)')
plt.ylabel('Amplitude')
plt.grid(True)

# Plot system output
plt.subplot(4, 1, 3)
plt.plot(n, y, label='System Output', color='blue')
plt.title('System Output Signal')
plt.xlabel('n (samples)')
plt.ylabel('Amplitude')
plt.grid(True)

# Combined plot: Clean input, Noisy input, and System output
plt.subplot(4, 1, 4)
plt.plot(n, x_clean, label='Clean Input Signal', color='red')
plt.plot(n, x_noisy, label='Noisy Input Signal', color='green')
plt.plot(n, y, label='System Output', color='blue')
plt.title('Comparison of All Signals')
plt.xlabel('n (samples)')
plt.ylabel('Amplitude')

# Add the legend to the fourth plot
plt.legend(loc='lower right')  # Location of the legend
plt.grid(True)

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```
