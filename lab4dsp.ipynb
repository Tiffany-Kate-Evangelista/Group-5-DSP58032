```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#N = np.linspace(-10, 10, 21)
N = np.arange(-10, 10)

plt.title('Periodic Signal')

def step(N):
	return np.where(N <= 0, 1, 0)

plt.stem(N, step(N), linefmt='b')
plt.grid()
plt.show()
```

```python
N = np.linspace(-2*np.pi, 6*np.pi, 500)

plt.figure(figsize=(9,5))
plt.title('Amplitude Shift')
plt.plot(N,np.sin(N), label=r'$\sin \theta$', color = 'blue')
plt.plot(N,np.sin(2*N), label=r'$\sin 2\theta$', color = 'red')

plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('Frequency Shift')
plt.plot(N,np.sin(N), label=r'$\sin \theta + \frac{\pi}{4}$', color = 'blue')
plt.plot(N,np.sin(4*N), label=r'$\sin 4\theta$', color = 'red')

plt.legend()
plt.grid()
plt.show()
```

```python
N = np.linspace(-10, 10, 21)

def step(N):
	return np.where(N >= 0, 1, 0)

plt.stem(N, step(N))

plt.grid()
plt.show()
```

```python
N = np.linspace(-2*np.pi, 2*np.pi, 100)

plt.figure(figsize=(10,4))
plt.title('Amplitude Shift')
plt.plot(N,np.sin(N), label=r'$\sin \theta$')
plt.plot(N,np.sin(2*N), label=r'$\sin 2\theta$')

plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.title('Frequency Shift')
plt.plot(N,np.sin(N), label=r'\sin \theta')
plt.plot(N,np.sin(2*N), label=r'$\sin 2\theta$')

plt.legend()
plt.grid()
plt.show()
```
