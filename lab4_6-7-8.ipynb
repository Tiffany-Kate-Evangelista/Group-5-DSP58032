```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N = np.linspace(-10, 10, 9)

def impulse(N):
    return np.where(N == 2.5, 1, 3)
plt.title("Code 6: Impulse Signal")
plt.stem (N, impulse(N))
plt.grid()
plt.show()
```

```python
def ramp (t, m, b):
    return m*t + b

plt.plot (N, ramp(N, 1, -2), label = "m=1, b=-2", color = "yellow")
plt.plot (N, ramp(N, -1, 2), label = "m=-1, b=2", color = "red")
plt.plot (N, ramp(N, 1, 2), label = "m=1, b=2", color = "green")
plt.plot (N, ramp(N, -1, -2), label = "m=-1, b=-2", color = "blue")

plt.axhline(color="k")
plt.axvline(color="k")

plt.title("Code 7: Ramp Signal")
plt.legend()
plt.grid()
plt.show()
```

```python
N = np.linspace(-2, 2, 35)

plt.stem(N, np.exp(-N-1), linefmt = "blue", label = "$e^{-n-1}$")
plt.stem(N, np.exp(-N+1), linefmt = "violet", label = "$e^{-n+1}$")
plt.stem(N, np.exp(N-1), linefmt = "dodgerblue", label = "$e^{n-1}$")
plt.stem(N, np.exp(N+1), linefmt = "pink", label = "$e^{n+1}$")

plt.title("Code 8: Exponential Signal")
plt.grid()
plt.legend()
plt.show()
```
